Let's check about how we select rows and columns using iloc and loc
* iloc = integer location
* loc = location

```python
# iloc for integer position, filtering rows and selecting columns by integer position
# iloc is exclusive of column/row range 0-2, iloc stops at 0-1
#-------------------------------------------------------------------------------------------------------
# loc for labels, filtering rows and selecting columns
# loc is inclusive of column/row range 0-2, starts with 0 ends with 2
#-------------------------------------------------------------------------------------------------------
# iloc similar to iteration by number 
# loc similar to iteration by element
```

```python
# double paranthesis indicate that specific elements only and not the range 
df.iloc[[0,5,6]] 
```

```python
# df.iloc[rows : columns] 
# 0 = Temp, 1 = Weaather, 2 = Humid
df.iloc[1:3,1] 
```

```python
df.iloc[[True,True,False,True]] 
```

```python
df.head(11) 
```

```python
# Let's start with loc
# df.loc[df['Weaather'] == 'Windy' ,:]
df.iloc[df['Weaather'] == 'Windy' ,:] # iloc gives an error
```

```python
# here you get all values from 0 to 6 and not 5
df.loc[[0,6], :]
#df.loc[1:3, :]
```

```python
df.loc[df.Humid > 50] 
```

```python
df.loc[(df.Humid > 50) & (df.Humid < 90),:]   #AND OPERATOR
#df.loc[(df.Humid > 60) | (df.Humid == 52),:]   #OR OPERATOR
```
